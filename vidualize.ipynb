```python
%matplotlib inline
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import glob
```

```python
import matplotlib as mpl
print(mpl.matplotlib_fname())
```

```python
savepath =  "./trypsin28_43_precon1_6mat52_init1"
results = sorted(glob.glob(savepath + "/csv/*"))
results
```

```python
for datapath in results:
    df = pd.read_csv(datapath, index_col=0)
    data = df.iloc[:,2:]
    pg = sns.pairplot(data)
    axes = pg.axes
    axes[0,0].set_ylim(-0.2,1.2)
    axes[1,0].set_ylim(0,30)
    axes[2,0].set_ylim(1,510)
    axes[3,0].set_ylim(0,6.5)
    axes[4,0].set_ylim(0,110)
    axes[5,0].set_ylim(-1,20)
    axes[6,0].set_ylim(-1,20)

    axes[0,0].set_xlim(-0.2,1.2)
    axes[0,1].set_xlim(0,30)
    axes[0,2].set_xlim(1,510)
    axes[0,3].set_xlim(0,6.5)
    axes[0,4].set_xlim(0,110)
    axes[0,5].set_xlim(-1,20)
    axes[0,6].set_xlim(-1,20)
    
    name = os.path.splitext(os.path.basename(datapath))[0]
    pg.savefig(savepath + "/fig/"+name+".png")
```
